```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['fueltype']=le.fit_transform(df['fueltype'])
df['aspiration']=le.fit_transform(df['aspiration'])
df['enginelocation']=le.fit_transform(df['enginelocation'])
df['fuelsystem']=le.fit_transform(df['fuelsystem'])
df['enginetype']=le.fit_transform(df['enginetype'])
df['drivewheel']=le.fit_transform(df['drivewheel'])
df['carbody']=le.fit_transform(df['carbody'])
df['fuelsystem']=le.fit_transform(df['fuelsystem'])
```

```python
df['cylindernumber'].unique()
```

```python
df.dtypes
```

```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
y
```

#train the model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

#model selection

1.Linear Regressio: Use it when the relationship between inputs and output is straight-forward and simple.


2.Support Vector Regression (SVR): Good for complex, non-linear data where you need more flexibility and donâ€™t mind extra tuning.


3.Decision Tree Regression: there are 10 category columns so we will be using decision tree model

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

# Linear Regression

```python
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print(f'mean_square_error:', mse)
print(f'mean_absolute_error:', mae)
print(f'r2_score:', r2)
```

# SVR

```python
svr=SVR()
```

```python
param={'kernel':['linear','poly','rbf','sigmoid']}
```

```python
gs=GridSearchCV(SVR(),param_grid=param)
```

```python
gs.fit(x_train,y_train)
```

```python
gs.best_estimator_.fit(x_train,y_train)
```

```python
y_pred=gs.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f'mean_square_error:', mse)
print(f'mean_absolute_error:', mae)
print(f'r2_score:', r2)
```

# decision tree

```python
DS=DecisionTreeRegressor()
```

```python
DS.fit(x_train,y_train)
```

```python
y_pred=DS.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f'mean_square_error:', mse)
print(f'mean_absolute_error:', mae)
print(f'r2_score:', r2)
```

# conclusion

among the algorithm  we have trained decision tree regressor gives the best score so we will go with decision tree regressor for this data
